```python
__author__ = 'Murtuza'
import requests
import csv
import re
import unicodecsv
from bs4 import BeautifulSoup
import socks
import socket

#socks.set_default_proxy(socks.SOCKS5, "localhost", 9150)
#socket.socket = socks.socksocket
page_count =12

session = requests.Session()
hdr = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/46.0.2490.80 Safari/537.36',
                       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8','Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
                       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
                       'Accept-Encoding': 'none',
                       'Accept-Language': 'en-US,en;q=0.8',
                       'Connection': 'keep-alive'}

outputFile = open("output_Nykaa_reviews_2.csv", 'wb')

writer = unicodecsv.writer(outputFile)
writer.writerow(('product_name', 'product_id','review_id', 'review_author', 'review_title','review_text', 'stars','date'))
productlistFile = open("output_Nykaa_Product_list.csv", 'r')
#print productlistFile
next = productlistFile.readline()
print next
count = 0
productlist = []
product_count = 0
while next != "": 
    productlist.append(next)
    next = productlistFile.readline()
    #print productlist
for a in productlist:
    product_name = a.strip("\r\n")
    print product_name
    page_getter = "http://www.nykaa.com/" + str(product_name)+ ".html?p=1&req=rev&sort=newest&filter=all&p=&isAjax=1&req=rev&sort=newest&filter=all"
    print page_getter
    product_count = product_count + 1
    print "Product Number: %d"  %(product_count)
    req = session.get(page_getter, headers = hdr)
    soup = BeautifulSoup(req.text)
    soup_str = str(soup)
    #print soup
    if "totalReview" in soup_str:
        page_count_id = soup.find("input",{"id":"totalReview"})  #counting the number of pages for this product
        review_count = int(page_count_id.get("value"))
        page_numbers = range(1,page_count+1)
        No_of_pages = len(page_numbers)
        print "No. of Pages: %d"   %(page_count)  #No. of review pages for this product
        for i in page_numbers:
            print i
            url = "http://www.nykaa.com/" + product_name+".html?p="+str(i)+"&req=rev&sort=newest&filter=all&p="+str(i)+"&isAjax=1&req=rev&sort=newest&filter=all"
            print url
            request = session.get(url, headers = hdr)
            bs0bj = BeautifulSoup(request.text)
            product_id = bs0bj.find("input",{"id":"product_conf_id"})
            product_id = product_id.get("value")
            print product_id
            htmltext = bs0bj.findAll("div",{"class":"product-tab-reviews"})
            No_of_reviews = len(htmltext)  #no. of reviews given by Nykaa server for this page
            print "No. of Reviews given by Nykaa server for this page: %d" %(No_of_reviews)
            #print htmltext
            counter = 0
            for x in htmltext:
                product_name_holder = product_name
                product_id_holder = product_id
                review_id_stars = x.find("div",{"class":"get_dynamic_rating"})
                review_id = review_id_stars.get("id")
                review_stars = review_id_stars.get("rating")
                #print review_id
                #print review_stars
                review_user_name = x.find("div",{"class":"r-name"})
                reviewer_name = review_user_name.find("p").get_text()
                #print reviewer_name
                review_title = x.find("p",{"class":"r-heading"}).get_text()
                #print review_title
                review_text = x.find("p",{"style":"margin: 1em 0;clear: both;"}).get_text()
                #print review_text
                review_date = x.find("p",{"class":"posted-date"}).get_text()
                review_date = re.findall(r'\d*/\d*/\d*', review_date)
                review_date = review_date[0]
                #print review_date
                writer.writerow((product_name_holder, product_id_holder, review_id, reviewer_name, review_title,review_text, review_stars,review_date))
                #if counter == len(page_numbers_list):
                    #break
                #counter = counter +1
                #print counter
    else:
        print "No reviews for this product: %s" %(product_name)

outputFile.close()
productlistFile.close()


        
        
```
